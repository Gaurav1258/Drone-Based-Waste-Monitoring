```python
data = []
train = "/content/train"
import os
for filename in os.listdir(train):
    if filename.endswith("jpg"):
        # Your code comes here such as
        print(filename)
        data.append(filename)
```

```python
from PIL import Image

model = tf.keras.models.load_model('/gdrive/My Drive/IOT/IOT_Waste_Model.h5')
predictions = []

for filename in os.listdir(train):
    img = Image.open(os.path.join(train, filename))
    i = "/content/train/"+filename
    print(i)
    img = tf.keras.preprocessing.image.load_img(i, target_size=(256, 256))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)
    pred = model.predict(img_array)
    class_label = np.argmax(pred)
    predictions.append(class_label)
    img.close()

print(predictions)
```

```python
pred_arr = np.array(predictions)
```

```python
filename_arr=np.array(data)
```

```python
# concatenate the arrays column-wise
output = np.column_stack((filename_arr, pred_arr))
print(output)
```

```python
import pandas as pd

df = pd.DataFrame({'classes': predictions})

# create a dictionary to map integer values to categorical strings
classes_mapping = {1: 'Aluminium', 2: 'Carton', 3: 'Glass', 4:'Organic Waste', 5:'Other Plastics', 6:'Paper and Cardboard',
                        7:'Plastic', 8:'Textiles', 9:'Wood'}

# use the map() function to apply the mapping to the dataframe column
df['classes_mapping'] = df['classes'].map(classes_mapping)

print(df)
```

```python
class_arr = df['classes_mapping'].to_numpy()
```
